**1. Processing**

*a. tampilkan ringkasan stasistik*

```python

import pandas as pd
df = pd.read_excel("data_pelanggan.xlsx")

print("Data Teratas:")
print(df.head())

print("\nRingkasan Statistik:")
print(df.describe())
```

*b. Visualisasikan hubungan antara pendapatan dan pembelian_tahunan*

```python
# Import library visualisasi
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot menggunakan matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(df['pendapatan'], df['pembelian_tahunan'], alpha=0.7, color='blue')
plt.title('Hubungan antara Pendapatan dan Pembelian Tahunan', fontsize=14)
plt.xlabel('Pendapatan', fontsize=12)
plt.ylabel('Pembelian Tahunan', fontsize=12)
plt.grid(True)
plt.show()

# Scatter plot menggunakan seaborn (opsional)
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='pendapatan', y='pembelian_tahunan', data=df, color='green')
plt.title('Hubungan antara Pendapatan dan Pembelian Tahunan (Seaborn)', fontsize=14)
plt.xlabel('Pendapatan', fontsize=12)
plt.ylabel('Pembelian Tahunan', fontsize=12)
plt.show()
```

**2. Implementasi Algoritma**

*a. Decission Tree*

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

df['label'] = df['umur'].apply(lambda x: 1 if x > 0 else 0)
features = df[['pembelian_tahunan', 'pendapatan',]].fillna(0)
labels = df['label']

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
